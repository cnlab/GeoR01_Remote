```python
outlier = []

for f in glob.glob(os.path.join(fmriprep_path, 'sub-*/func/*confounds_regressors.tsv')):
    print("*", end="")
    
    bn = os.path.basename(f)
    sub = bn.split('sub-')[1].split("_")[0]
    task = bn.split('task-')[1].split("_")[0]
    run = int(bn.split('run-')[1].split("_")[0])
    
    confounds = pd.read_csv(f, sep="\t")
    
    gs = (np.abs(scale(confounds['global_signal'])) > 3)
    fd = (confounds['framewise_displacement'] > 0.75)
    gs_fd = np.any([gs, fd], axis=0)
    
    outlier.append({"sub": sub, "task": task, "run": run, "gs": np.mean(gs), "fd": np.mean(fd), "gs_fd": np.sum(fd)})

outlier = pd.DataFrame(outlier)
```
